```python
## Checking on the correlation between the variables
Cor=df.iloc[:,0:6].corr(method='pearson')
plt.figure(figsize=(12 ,8))
mask = np.array(Cor)
mask[np.tril_indices_from(mask)] =False
sns.heatmap(data =Cor, annot =True, mask =mask,vmax =1, vmin=-1, cmap ='coolwarm');
```

```python
## Check for Outliers
# construct box plot for continuous variables
cont=df.dtypes[(df.dtypes!='uint8') & (df.dtypes!='bool')].index
plt.figure(figsize=(10,10))
df[cont].boxplot(vert=0)
plt.title('With Outliers',fontsize=16)
plt.show()
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in df[cont].columns:
    lr,ur=remove_outlier(df[column])
    df[column]=np.where(df[column]>ur,ur,df[column])
    df[column]=np.where(df[column]<lr,lr,df[column])
```

```python
plt.figure(figsize=(10,10))
df[cont].boxplot(vert=0)
plt.title('After Outlier Removal',fontsize=16)
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
X = df[['R','AB','H','2B','3B','HR','BB','SO','SB','RA','ER','ERA','CG','SHO','SV','E']]
vif = pd.DataFrame()
vif["Variables"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
import statsmodels.formula.api as smf
def vif_cal(input_data):
  x_vars = input_data
  xvars_names = input_data.columns
  for i in range(0,xvars_names.shape[0]):
    y=x_vars[xvars_names[i]]
    x=x_vars[xvars_names.drop(xvars_names[i])]
    rsq = smf.ols(formula ="y~x", data =x_vars).fit().rsquared
    vif = round(1/(1-rsq),2)
    print(xvars_names[i], "VIF = ", vif)
```

```python
## Ouput using from statsmodels.stats.outliers_influence import variance_inflation_factor
print(vif)
```

```python
## Now dataset is ready for linear regression.We seperate the dependent and independent variables
# Copy all the predictor variables into X dataframe
X = df.drop('W', axis=1)

# Copy target into the y dataframe. 
y = df[['W']]
```

```python
X.head()
```

```python
# Split X and y into training and test set in 75:25 ratio
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25 , random_state=1)
```

# Linear Regression Model

```python
# invoke the LinearRegression function and find the bestfit model on training data
## We fit regression model on training data
## LinearRegression() -  We create an object, that is, we invoke the linear regression function

regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
# Let us explore the coefficients for each of the independent attributes
## Enumerate means -  show me column names one by one
## idx -  Index
## regression_model.coef_[0][idx] - if you use this code and if you are getting same coeffcient, then remove [0] or use regression_model.coef_[idx]

for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model.coef_[0][idx]))
```

```python
## The output is a 2-diamentional array - see 2 square brackets

regression_model.coef_
```

```python
## One diamentional array
regression_model.coef_[0]
```

```python
# Let us check the intercept for the model

intercept = regression_model.intercept_[0]

print("The intercept for our model is {}".format(intercept))
```

```python
# R square on training data
## The score here is the coefficient of determination
regression_model.score(X_train, y_train)
```

```python
# R square on testing data
regression_model.score(X_test, y_test)
```

```python
#RMSE on Training data
predicted_train=regression_model.fit(X_train, y_train).predict(X_train)
np.sqrt(metrics.mean_squared_error(y_train,predicted_train))
```

```python
#RMSE on Testing data
predicted_test=regression_model.fit(X_train, y_train).predict(X_test)
np.sqrt(metrics.mean_squared_error(y_test,predicted_test))
```
